# Chapter 7 - Deploy to Production

### Mobile

```
import torch
import torchvision
from torch.utils.mobile_optimizer \
  import optimize_for_mobile

model = torchvision.models.vgg16(pretrained=True)
model.eval()
example = torch.rand(1, 3, 224, 224) # <1>

traced_script_module = \
  torch.jit.trace(model, example) # <2>
torchscript_model_optimized = \
  optimize_for_mobile(traced_script_module) # <3>
torchscript_model_optimized.save("model.pt") # <4>
```
